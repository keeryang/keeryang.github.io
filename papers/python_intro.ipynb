Step 5. An extensible and reuseable program!

```python
def datadescribe(df_input, y = 0,var_x = ['EMPLOYEESALLSITES'], gp = 'CITYCODE'):
    # Either on all data, or just one year data
    if y == 0:
        df = df_input
    else:    
        df = df_input.loc[df_input['YEAR'] == y]
    #Step 1. Show the data?
    print("Step 1. Reading the data")
    print(df.head(5))
    #Step 2. What data is actually available?
    print("Step 2. What data is actually available?")
    print(df.isnull().sum().sort_values(ascending=True)/len(df.index))
    #Step 3. Describing the data
    print("Step 3. Describing the data")
    df[var_x].describe()
    df[var_x].corr()
    df[var_x].hist(figsize=(5,4))
    plt.show()
    #Step 4. Answer the original question
    print("Step 4. Answer the original question")
    print(df[var_x + gp].groupby(gp).describe())
```

```python
datadescribe(df_input = df, y = 0,var_x = ['EMPLOYEESALLSITES'], gp = ['CITYCODE'])
```

```python
datadescribe(df_input = df, y = 0,var_x = ['EMPLOYEESALLSITES','SLS'], gp = ['CITYCODE'])
```

```python
datadescribe(df_input = df, y = 2000,var_x = ['EMPLOYEESALLSITES'], gp = ['SIC2'])
```
